#### Range


The range() function returns a sequence of numbers, starting from 0 by default, and increments by 1 (by default), and stops before a specified number.

For example if you use list(range(10)) it will generate a list for you starting from 0 - 9

```python
# Assings value of 10 in the range function starting form 0 - 9
range(10)
```

```python
# Creats a list with values from 0 to 9
list(range(10))
```

```python
# Creats a list with values from 3 to 9 (3 is the starting point and 9 the end point)
list(range(3, 10))
```

```python
# Creats a list with values from 10 to 3 (10 is the starting point and 3 the end point) For this you have to instert a stepzise of -1
list(range(10, 3, -1))
```

```python
# range function does not work with float numbers
list(range(10, 3.5, -1))
```

#### For Statement




A for loop is used for iterating over a sequence (that is either a list, a tuple, a dictionary, a set, or a string).

```python
l = [3, 4, 5, 6, 7, 8, 23, 4, 5, 5]
```

```python
# extract all the data from the list l
l = [3, 4, 5, 6, 7, 8, 23, 4, 5, 5]
for i in l:
    print(i)
```

```python
# Adds + 2 to all the values in the list l to the created lis m
l = [3, 4, 5, 6, 7, 8, 23, 4, 5, 5]
m = [] # created an empty list for the new values as input

for i in l:
    m.append(i+2)
```

```python
# as you can see the above function created a new list m and added l values + 3
print(l)
print(m)
```

```python
# Adds + 2 to all the values in the list to a new list with the use of range

l = [3, 4, 5, 6, 7, 8, 23, 4, 5, 5]
m = []

for i in range(len(l)):
    m.append(l[i]+2)
```

```python
# Adds + 2 to all the values in the list to a new list with the use of if statement

l = [3, 4, 5, 6, 7, 8, 23, 4, 5, 5]
m = []

for i in l:
    if i >=10:
        m.append(i+2)
```
