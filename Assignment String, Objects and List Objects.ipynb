#### Write a Python program to reverse a word after accepting the input from the user.

**Sample Output**:
    
input word: Prashant\
Output: tnahsarP

```python
word = input("Input a word to reverse: ")

for char in range(len(word) - 1, -1, -1):
    print(word[char], end="")
print("\n")
```
