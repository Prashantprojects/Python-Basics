**lambda**

A lambda function is a small anonymous function.

A lambda function can take any number of arguments, but can only have one expression.

```python
l = [3,4,5,6,7,8,9,9]
```

```python
# values squared 
list(map(lambda x: x **2,l))
```

```python
# values squared 0
l = [3,4,5,6,7,8,9,9]
c = lambda x: x **2

list(map(c,l))
```

**reduce()**

Python's reduce() is a function that implements a mathematical technique called folding or reduction. reduce() is useful when you need to apply a function to an iterable and reduce it to a single cumulative value.

```python
from functools import reduce
```

```python
l = [4,5,6,7,8,9]

def summation(a,b):
    return a +b
```

```python
l = [4,5,6,7,8,9]

def summation(a,b):
    return a +b
```

```python
l = [4,5,6,7,8,9]

reduce(summation,l)
```

```python
l = [4,5,6,7,8,9]

reduce(lambda x,y: x+y,l)
```

```python
 l = [4,5,6,7,8,9]

reduce(lambda x,y: x/y,l)
```

```python
 l = [4,5,6,7,8,9]

reduce(lambda x,y: x*y,l)
```

**filter()**

```python
# filter out all the even data from the list
l = [4,5,6,7,8,6,7,8,8]

list(filter(lambda x: x%2 ==0,l))
```

```python
# filter out all the uneven data from the list
l = [4,5,6,7,8,6,7,8,8]

list(filter(lambda x: x%2 !=0,l))
```

```python
def even_check (n):
    if n%2 == 0:
        return True
```

```python
list(filter(even_check,l))
```

```python
def even_check (n):
    if n%2 != 0:
        return True
```

```python
list(filter(even_check,l))
```
