#### Write a program which wil find all such numbers which are divisble by 7 but are not a mulitple of 5, between 2000 and 3200 (both included). The numbers obtained should be printed in a comma-seperated sequance on a single line

```python
# create a empty list
l = []

# create a loop
for x in range(2000, 3200):
     if(x%7==0) and (x%5==0):
        l.append(str(x))
print(','.join(l))
```
