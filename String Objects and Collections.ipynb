**extend()**

```python
# append (add item to list)
s = [1,3,4,5,5]
m = [4,5,6,7,"prash"]

s.append([3,4,5,6,7])

# creates a new list with the new items
m.extend([6,7,8,"bhiekhemsing"])

print(s)
print(m)
```
